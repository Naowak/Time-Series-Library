```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example memory footprints in MB for different models
footprints = {
    'EST': 7409,
    'Transformer': 979,
    'Reformer': 1373,
    'iTransformer': 1679,
    'PatchTST': 4157,
    'TimesNet': 1025
}

# Set the style for scientific papers
sns.set(style="whitegrid", context="paper", font_scale=1.2)

# Create the bar plot
plt.figure(figsize=(7, 4))
ax = sns.barplot(x=list(footprints.keys()), y=list(footprints.values()), palette="muted")

# Add value labels on top of bars
for i, v in enumerate(footprints.values()):
    ax.text(i, v + 100, f"{v}", ha='center', va='bottom', fontsize=10)

# Set labels and title
ax.set_xlabel("Model", fontsize=12)
ax.set_ylabel("Memory Footprint (MB)", fontsize=12)
ax.set_title("Peak Memory Footprints for 1M Parameters Models", fontsize=14, weight='bold')

# Remove top and right spines for a cleaner look
sns.despine()

plt.tight_layout()
plt.show()
```
