```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example memory footprints in MB for different models
footprints = {
    'PatchTST': 8615,
    'EST': 7035,
    'iTransformer': 1251,
    'Reformer': 1035,
    'TimesNet': 817,
    'Transformer': 611,
    'Mamba': 343,
}

# Set the style for scientific papers
sns.set(style="whitegrid", context="paper", font_scale=1.2)

# Create the bar plot
plt.figure(figsize=(7, 4))
ax = sns.barplot(x=list(footprints.keys()), y=list(footprints.values()), palette="muted")
# subtitle
ax.text(0.5, 1.03, "(Measured while training on a NVIDIA Quadro RTX 5000)", ha='center', va='center', fontsize=12, transform=ax.transAxes)

# Add value labels on top of bars
for i, v in enumerate(footprints.values()):
    ax.text(i, v + 100, f"{v}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Set labels and title
# ax.set_xlabel("Model", fontsize=12)
ax.set_ylabel("Memory Footprint (MB)", fontsize=12, weight='bold', labelpad=15)
ax.set_title("Peak Memory Footprints for 1M Parameters Models", fontsize=14, weight='bold', pad=20)

# Remove top and right spines for a cleaner look
sns.despine()

plt.tight_layout()
plt.savefig("memory_footprint.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
